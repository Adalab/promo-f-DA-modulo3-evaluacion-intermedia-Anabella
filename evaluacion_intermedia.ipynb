Preparaci√≥n de datos 

```python
#Todos los nombres de las columnas en minuscula 
columnas_minusculas = {columna: columna.lower() for columna in df_spotify.columns}
```

```python
columnas_minusculas
```

```python
#Luego del for renombramos las columnas 
df_spotify.rename(columns=columnas_minusculas, inplace= True)
```

```python
#df.reindex(lista de columnas) para cambiar el orden de las columnas
#Reccordar que las columnas siemre se pueden pasar a lista y aplicar metodo
df_spotify.columns
```

```python
columnas_modificar = ['danceability','key','energy', 'loudness', 'speechiness', 'acousticness','instrumentalness']
```

```python
#Reemplazamos las comas por los puntos en las columnas: (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness)
def limpiar_datos (columna):

     try:
          df_spotify[columna] = (df_spotify[columna].str.replace(',', '.')) 
          return df_spotify[columna]
     except:
          np.nan
```

```python
for columnas in columnas_modificar:
    df_spotify[columnas]=limpiar_datos(columnas)
    print(df_spotify[columnas])
```

```python
#Creamos una funcion para pasar los valores de object a float 
def conversion (valores):
    try:
        for columna in columnas_modificar:
            df_spotify[columna]= pd.to_numeric(df_spotify[columna]).astype(float)
    except:
        pass
```

```python
#llamams a la funcion 
conversion(columnas_modificar)
```

```python
#Comprobamos las modificacion de los datos
df_spotify.info()
```

```python
columnas_convertibles = ['title', 'album', 'track']
for col in columnas_convertibles:
    df_spotify[col] = df_spotify[col].apply(lambda x: x.lower() if type(x) == str else x)
```

```python
df_spotify.head(3)
```

```python
#Funcion para cambiar el signo $ de la columna de artistas por una s
def artistas (valores):
    return  valores.replace('$', 's')
     
```

```python
df_spotify['artist']=df_spotify['artist'].apply(artistas)
```

Usando el dataframe de los albumes que hemos creado en el ejercicio anterior. ¬øCu√°l es la media y la desviaci√≥n est√°ndar de "danceability", "acousticness" y "speechiness" de cada artista?. Muesrtra los resultados en tres dataframes diferentes, uno para cada m√©trica ("danceability", "acousticness" y "speechiness"). Ordena los resultados de mayor a menor en base a la media.

```python
df_danceability = df_album.groupby('artist')['danceability'].agg(['mean', 'std']).reset_index().sort_values(by= 'mean', ascending=False)
df_danceability
```

```python
df_acousticness = df_album.groupby('artist')["acousticness" ].agg(['mean', 'std']).reset_index().sort_values(by= 'mean', ascending=False)
df_acousticness
```

```python
df_speechiness = df_album.groupby('artist')["speechiness" ].agg(['mean', 'std']).sort_values(by= 'mean', ascending=False).reset_index()
df_speechiness
```

```python
#Para todas las columnas juntas
metricas = ["danceability", "acousticness" y "speechiness"]
for met in metricas:
    df_art = df_album.groupby('artist')[met].agg(['mean', 'std']).reset_index().sort_values(by= 'mean', ascending=False)
    display(df_art)
```
